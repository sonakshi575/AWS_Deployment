```python
def weight_cal_month1(a,b):
    baby_1=a
    baby_2=b
    if baby_1[1]>baby_2[1]:
        print("Baby_1 weighed more than baby_2 on month 1")
    else:
        print("Baby_1 weighed less than Baby_2 on Month 1")
        

def weight_cal_last_month(a,b):
    baby_1=a
    baby_2=b
    if baby_1[11]>baby_2[11]:
        print("Baby_1 remained heavier than Baby_2")
    else:
        print("Baby_2 became heavier than Baby_1")
```
